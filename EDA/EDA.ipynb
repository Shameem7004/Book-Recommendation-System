## Data Cleaning

```python
# Get all User-IDs that are valid (i.e., exist in the user_df)
valid_user_ids = user_df['User-ID'].unique()
length = len(valid_user_ids)
length
```

```python
# Filter rating_df to only include ratings from these valid users
rating_df_cleaned = rating_df[rating_df['User-ID'].isin(valid_user_ids)]
rating_df_cleaned
```

```python
# Now, filter user_df to only include users who have actually rated something
active_user_ids = rating_df_cleaned['User-ID'].unique()
user_df_cleaned = user_df[user_df['User-ID'].isin(active_user_ids)]
user_df_cleaned
```

```python
## Calculate Average Rating for Each Unique ISBN

# Group the rating_df by 'ISBN' and calculate the mean of 'Book-Rating' for each ISBN and then resetting the index.
average_ratings = rating_df_cleaned.groupby('ISBN')['Book-Rating'].mean().reset_index()

# renaming the column.
average_ratings.rename(columns={'Book-Rating': 'Avg-Book-Rating'}, inplace=True)
average_ratings
```

```python
## Counts how many unique ISBNs are present in average_ratings.
average_ratings['ISBN'].nunique()
```

```python
merged_df = average_ratings.merge(book_df, on = 'ISBN')
merged_df
```

```python
## just to check exactly same book
same_book_merged_df = merged_df[merged_df.duplicated(subset=['Book-Title', 'Book-Author', 'Year-Of-Publication', 'Publisher'], keep=False)] 
same_book_merged_df
```

```python
# Define the new, more specific columns to identify a unique book
# A book is a duplicate only if Title, Author, Year, AND Publisher are all the same.
grouping_columns = [
    'Book-Title',
    'Book-Author',
    'Year-Of-Publication',
    'Publisher'
]

# Define the rules for how to combine the duplicate rows
# We will take the MEAN of the ratings and just the FIRST of the other info.
aggregation_rules = {
    'Avg-Book-Rating': 'mean',
    'Image-URL-M': 'first',
    'ISBN': 'first'
}

final_merged_book_df = merged_df.groupby(grouping_columns, as_index=False).agg(aggregation_rules)
final_merged_book_df
```

```python
# Get the current list of columns
current_columns = final_merged_book_df.columns.tolist()

# Remove 'ISBN' from its current position and insert it at the beginning
current_columns.insert(0, current_columns.pop(current_columns.index('ISBN')))

final_merged_book_df = final_merged_book_df[current_columns]
final_merged_book_df
```
